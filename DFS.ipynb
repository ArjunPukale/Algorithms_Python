```python
adj_list = {}
adj_list[1] = [4,5]
adj_list[2] = [3,4,6]
adj_list[3] = [2]
adj_list[4] = [1,2,5]
adj_list[5] = [1,4,6]
adj_list[6] = [2,5]
for i,j in adj_list.items():
    print(f"{i} --> {j}")
```

```python
visited = [0 for i in range(len(adj_list))]
parent = [0 for i in range(len(adj_list))]
```

```python
def DFS(i): # i-> start node
    visited[i-1] = 1
    print(i)
    for j in adj_list[i]:
        if not visited[j-1]:
            parent[j-1] = i
            DFS(j)
```
