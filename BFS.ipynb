## Creating Queue Class

```python
class Queue:
    def __init__(self):
        self.L = []   # [1,2,3 ..... n]<---tail
    def isempty(self):
        return len(self.L) == 0
    def add(self,x):
        self.L.append(x)
    def extract_head(self):
        if self.isempty():
            return None
        else:
            j = self.L[0]
            self.L = self.L[1:]
            return j
    def display(self):
        print(self.L)
```

## Adjacency List

```python
adj_list = {}
adj_list[1] = [4,5]
adj_list[2] = [3,4,6]
adj_list[3] = [2]
adj_list[4] = [1,2,5]
adj_list[5] = [1,4,6]
adj_list[6] = [2,5]
for i,j in adj_list.items():
    print(f"{i} --> {j}")
```

## BFS Function to reach target from start node

```python
def BFS(i, target): # i-> starting node
    found_target = 0
    visited = [0 for i in range(1,len(adj_list)+1)]
    parent = [0 for i in range(1,len(adj_list)+1)]
    q = Queue()
    #start exploration at i
    visited[i-1] = 1
    q.add(i)
    #Explore each vertex in Q
    while not q.isempty():
        j = q.extract_head()
        for k in adj_list[j]: #visit all neighbours of j
            if not visited[k-1]:
                q.add(k)
                visited[k-1] = 1
                parent[k-1] = j
                if k == target:
                    found_target = 1
                    break
        if found_target:
            break
    if found_target:
        child = target
        Par = parent[target-1]
        path = [child, Par]
        while path[-1] != i :
            path.append(parent[path[-1] - 1])
        print("Path: ",'->'.join(map(str,path[::-1])))
    else:
        print(f"No connection exists between {i} and {target}")
```
